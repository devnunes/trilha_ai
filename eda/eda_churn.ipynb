# Merging dataframes

```python
# Load customer dataset from csv
df_customers = pd.read_csv('./datasets/churn_customers.csv')

# Merging dataframes customer and services into a new one
df_temp = pd.merge(df_customers, df_services, on='customerID')

df_temp.info() 
```

```python
# Merging temp and contracts, using columns with divergent names
# Lets rename the column customerID of contracts to CustomerID
df_contracts.rename(columns={'customerID': 'CustomerID'}, inplace=True)
df_churn_temp = pd.merge(df_temp, df_contracts, left_on='customerID', right_on='CustomerID')

df_churn_temp.info()
```

```python
# Unifying three dataframes at once
df_churn = df_customers.merge(df_services, on='customerID').merge(df_contracts, left_on='customerID', right_on='CustomerID')

df_churn.info()
```

```python
 # Drop column
df_churn.drop(columns=['CustomerID'], axis=1, inplace=True)

df_churn.info()
```
