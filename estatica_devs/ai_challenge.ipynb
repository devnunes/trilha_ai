Based on the statistics module classes, use Pandas library to process to dictionary below:

- Calculate the sales mean
- Create a vertical bar graph showing the reference month and the value.
- Create a line graph showing the reference month and the value.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Dictionary with the data of the sales
dict_revenue = {
    'reference_date': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'value': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
sns.set_theme()
```

```python
df_revenue = pd.DataFrame.from_dict(dict_revenue)
df_revenue['reference_date'] = pd.to_datetime(df_revenue['reference_date'])
df_revenue.info()
```

```python
formatter = plt.ScalarFormatter()
formatter.set_scientific(False)
formatter.set_useOffset(False)
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title('Sales Revenue')
ax.set_ylabel('Revenue')
ax.set_xlabel('Date')
ax.yaxis.set_major_formatter(formatter)
ax.get_xaxis().set_major_formatter(mdates.DateFormatter('%m/%Y'))
df_revenue = df_revenue.sort_values('reference_date')

ax.bar(df_revenue['reference_date'], df_revenue['value'], width=30)
plt.xticks(df_revenue['reference_date'], rotation=90)
plt.show()
```

```python
formatter = plt.ScalarFormatter()
formatter.set_scientific(False)
formatter.set_useOffset(False)
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title('Sales Revenue')
ax.set_ylabel('Revenue')
ax.set_xlabel('Date')
ax.yaxis.set_major_formatter(formatter)
ax.get_xaxis().set_major_formatter(mdates.DateFormatter('%m/%Y'))
df_revenue = df_revenue.sort_values('reference_date')

ax.plot(df_revenue['reference_date'], df_revenue['value'])
plt.xticks(df_revenue['reference_date'], rotation=90)
plt.show()
```
